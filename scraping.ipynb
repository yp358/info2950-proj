```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
from sklearn.linear_model import LinearRegression, LogisticRegression
import duckdb
import requests
from bs4 import BeautifulSoup
import time
```

```python
url = 'https://billboardtop100of.com/1960-2/'

page = requests.get(url) 

soup = BeautifulSoup(page.text, 'html.parser') 
```

```python
soup.find_all("div", class_= "entry-content clear") 
```

```python
y = soup.find("h1", class_="entry-title").get_text()

# Initialize empty lists to store year, rank, artist, and song data
year = []
rank = []
artist = []
song = []

# Find all rows in the table
table_rows = soup.find_all('tr')

# Loop through each row and extract columns
for row in table_rows:
    columns = row.find_all('td')  # Find all <td> elements in the row

    if len(columns) == 3:  # Ensure there are 3 columns (rank, artist, song)
        ranks = columns[0].text.strip()  # Get the rank
        artists = columns[1].text.strip()  # Get the artist
        songs = columns[2].text.strip()  # Get the song

        # Append the year and the extracted data to the respective lists
        # year.append(y)
        rank.append(ranks)
        artist.append(artists)
        song.append(songs)
            
     
```

```python
data = {"Year": year "Rank": rank, "Artist":artist, "Song":song} 

df = pd.DataFrame(data) 

print(df)
```
